```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('NFLX.csv')
```

```python
df.head()
df.info()
```

```python
df.isnull().sum()
```

```python
df.dropna()
```

Line Graph

```python
plt.figure(figsize=(20,8))
plt.plot(df['Date'].head(18) ,df['Close'].head(18),)
plt.title("NETFLIX STOCK PRICE (Day vs Closing price)", fontweight='bold',fontsize =18)
plt.xlabel('Date')
plt.ylabel('Closing price')
plt.legend(['Line'])
plt.show()
```

Multiple line plot

```python
plt.figure(figsize=(20,8))
plt.title("NETFLIX STOCK PRICE ANALYSIS", fontweight='bold',fontsize =18)
plt.plot(df['Date'].head(18),df['Open'].head(18),label = "Open price",color='green')
plt.plot(df['Date'].head(18),df['High'].head(18),label = "High price",linestyle ='dashed', color='blue')
plt.plot(df['Date'].head(18),df['Low'].head(18),label = " Low price",linestyle ='dashed',color='red')
plt.plot(df['Date'].head(18),df['Close'].head(18),label = "Close price",color='black')
plt.show()
```

```python
plt.figure(figsize=(20,8))
plt.scatter(df['Volume'],df['Open'])
plt.title('Scatter plot',fontweight = 'bold',fontsize =18)
#plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(20,8))
plt.bar(df['Date'].head(20),df['Low'].head(20),alpha =0.6,color='green')
plt.title('Bar graph',fontweight = 'bold',fontsize =18)
plt.show()
```

```python
plt.figure(figsize=(20,8))
plt.hist(df['Open'], bins=30,color='red',edgecolor='black',alpha = 0.8)
plt.title('Histogram',fontweight = 'bold',fontsize =18)
plt.show()
```

```python
ax = plt.axes(projection ='3d')
ax.plot(df['Open'],df['High'],df['Low'])
plt.title('3D projection')
plt.show()
```

```python

plt.figure( figsize = ( 20,8)) 
plt.title('Time series plot')
sns.lineplot( x = df['Date'], 
			y = df['High'], 
			data = df, 
			) 

#plt.xlabel( 'Months of the year 2018')  
pos = [ '2018-01-01', '2018-02-01', '2018-03-01', '2018-04-01', 
	'2018-05-01', '2018-06-01', '2018-07-01', '2018-08-01', 
	'2018-09-01', '2018-10-01', '2018-11-01', '2018-12-01'] 
lab = [ 'Jan', 'Feb', 'Mar', 'Apr', 'May', 'June', 
	'July', 'Aug', 'Sept', 'Oct', 'Nov', 'Dec'] 

plt.xticks( pos, lab) 
plt.show()
```

```python


barwidth = 0.25
fig = plt.subplots(figsize=(12,8))
br1 = np.arange(len(df['Open'].head(5)))
br2 = [x+ barwidth for x in br1]
br3 = [x+ barwidth for x in br2]
plt.bar(br1, df['Open'].head(5),color ='r', width = barwidth, edgecolor = 'black',label= 'date' )
plt.bar(br1, df['High'].head(5),color ='g', width = barwidth, edgecolor = 'black',label= 'High' )
plt.bar(br1, df['Close'].head(5),color ='b', width = barwidth, edgecolor = 'black',label= 'close')
plt.xticks([r + barwidth for r in range(len(df['Open'].head(5)))])
plt.show()
```

```python
fig, ax = plt.subplots(4)
ax[0].plot(df['Date'].head(5),df['Open'].head(5),color = "red")
ax[1].plot(df['Date'].head(5),df['Close'].head(5),color = "green")
ax[2].plot(df['Date'].head(5),df['High'].head(5),color = "blue")
ax[3].plot(df['Date'].head(5),df['Low'].head(5),color = "violet")
plt.show()
```
