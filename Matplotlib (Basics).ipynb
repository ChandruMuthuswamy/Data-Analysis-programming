```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Normal graphs

```python
#Line graph - .plot
x = [1,2,3,4,5,]
y = [2,3,6,6,10]
plt.plot(x,y)
plt.title('Line Graph')
plt.legend(["Line"])
plt.show()
```

```python
#Bar graph - .bar

plt.bar(x,y)
plt.title("Bar graph")
plt.legend(['Bar'])
plt.show()
```

```python
#scattering - .scatter
plt.scatter(x,y)
plt.title("Scatter chart")
plt.legend(["Scatter"])
plt.show()
```

```python
#pie chart - .pie
size = [40,20,30,10]
l = ['python','c++','Java','Javascript']
e  =(0.2, 0, 0, 0)
plt.pie(size,labels = l)
plt.title("Pie chart")
plt.legend(["Pie chart"])
plt.show()
```

```python
#3D projection 
x=[1,2,3,4,5]
y =[1,4,20,16,25]
z =[1,8,27,64,125]
fig = plt.figure()
ax = plt.axes(projection ='3d')
ax.plot(x,y,z)
plt.show()
```

```python
x = [40,50,20,70,90]
plt.hist(x, bins = 30)
plt.title("Histogram")
plt.show()
```

```python
days = [1,2,3,4,5,6,7,8,9]
studying = [1,3,5,7,9,2,3,4,5]
playing =[2,4,6,8,2,3,5,6,9]
plt.xlabel('Days')
plt.ylabel('hours')
plt.stackplot(days,studying,playing,colors=['r','c'])
plt.show()
```

creating sub plots

```python
x = [1,2,3]
y=[2,4,3]
z=[4,1,2]
fig, ax = plt.subplots(3)
ax[0].plot(x,y,color = "red")
ax[1].plot(y,z)
ax[2].plot(x,z)
plt.show()
```

```python
fig ,ax = plt.subplots(2, 2)
ax[0,0].plot([1,2,5,9],[2,3,6,7])
ax[0,1].bar([2,4,6,9],[1,3,5,6])
ax[1, 0].bar([1, 2, 3], [4, 5, 6])
ax[1, 1].hist([1, 2, 2, 3, 3, 3, 3, 3, 3, 4, 4, 5])
ax[0, 0].set_title("Line")
fig.suptitle("Subplottings")
plt.show()
```

```python
# import required modules
import matplotlib.pyplot as plt 
import matplotlib.tri as mtri 
import numpy as np 
	
# assign data 
x = np.asarray([0, 1, 2, 3, 0.5, 
				1.5, 2.5, 1, 2, 
				1.5]) 

y = np.asarray([0, 0, 0, 0, 1.0, 
				1.0, 1.0, 2, 2, 
				3.0]) 

triangles = [[0, 1, 4], [1, 5, 4], 
			[2, 6, 5], [4, 5, 7], 
			[5, 6, 8], [5, 8, 7], 
			[7, 8, 9], [1, 2, 5], 
			[2, 3, 6]] 

# depict illustration
triang = mtri.Triangulation(x, y, triangles) 


fig, axs = plt.subplots() 
axs.tricontourf(triang,x) 
axs.triplot(triang, 'go-', color ='white') 

# turn off the axes
axs.set_axis_off()

# assign title
axs.set_title('Triangle illustration') 

plt.show() 
```
