```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
fruits = ['apple','orange','pineapple','banana','cherry']
sales = [200,150,125,250,50]
plt.xlabel('fruits')
plt.ylabel('sales')
plt.title('Fruit sales')
plt.bar(fruits,sales,color = "pink",width =0.3)
plt.show()
```

```python
plt.barh(fruits,sales,color = "green")
```

```python
barWidth = 0.25
fig = plt.subplots(figsize =(12, 8)) 

IT = [12, 30, 1, 8, 22] 
ECE = [28, 6, 16, 5, 10] 
CSE = [29, 3, 24, 25, 17] 

br1 = np.arange(len(IT)) 
br2 = [x + barWidth for x in br1] 
br3 = [x + barWidth for x in br2] 

plt.bar(br1, IT, color ='r', width = barWidth, edgecolor ='black', label ='IT') 
plt.bar(br2, ECE, color ='g', width = barWidth, edgecolor ='black', label ='ECE') 
plt.bar(br3, CSE, color ='b', width = barWidth, edgecolor ='white', label ='CSE') 

plt.xlabel('Branch', fontsize = 15) 
plt.ylabel('Students passed',  fontsize = 15) 
plt.xticks([r + barWidth for r in range(len(IT))], 
        ['2015', '2016', '2017', '2018', '2019'])

plt.legend()
plt.show() 
```

```python
cars = ['Tata','Renault','BMW','Skoda','Jeep']
sales = [200,150,125,250,50]
e = (0,0,0,0.1,0)
wp = {'linewidth': 1, 'edgecolor': "black"}
plt.title('car sales')
plt.pie(sales,labels=cars,explode =e,wedgeprops=wp,shadow =True,startangle=0, autopct = '%0.1f%%')
plt.savefig('firstpie1.png')
plt.style.use('classic')
plt.show()
```
